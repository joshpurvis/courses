```python
from math import sqrt
import itertools

with open("tsp.txt") as file:
    content = [l.split() for l in file.readlines()[1:]]
    content = [(float(x), float(y)) for [x,y] in content]

def distance(p1, p2):
    x1, y1 = p1
    x2, y2 = p2
    return sqrt((x1-x2)**2 + (y1-y2)**2)

distances = [[distance(pi,pj) for (j, pj) in enumerate(content)] for (i, pi) in enumerate(content)]
```

```python
g = {}

g[0] = {}
for i in range(1,25):
    g[0][(i, frozenset())] = distances[i][0]

g[1] = {}
for i in range(1,25):
    for j in range(1,25):
        if i != j:
            c = distances[i][j]
            g[1][(i, frozenset([j]))] = c + g[0][(j, frozenset())]
```

```python
combinations = itertools.combinations
def remove_one(subset):
    l = len(subset)
    for c in combinations(subset, l-1):
        diff = list(subset.difference(c))[0]
        yield (diff, frozenset(c))
        
def subresult(n, i, j, subset):
    return distances[i][j] + g[n][j, subset]
```

```python
for iteration in range(2,25):
    del g[iteration-2]
    g[iteration] = {}
    print(iteration)
    for c in map(frozenset, combinations(range(1,25), iteration)):
        for i in range(1,25):
            if i not in c:
                sub = min(subresult(iteration -1, i, j, s) for (j,s) in remove_one(c))
                g[iteration][i,c] = sub
```
